```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Read video
cap = cv2.VideoCapture("Traffic.mp4")

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Detect objects
    results = model(frame)

    # Draw detections
    annotated_frame = results[0].plot()

    cv2.imshow("YOLOv8 Detection", annotated_frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import torch

# Load YOLOv8 model
model = YOLO("yolov8x.pt")

# Move model to GPU
model.to("cuda")

# Open video
cap = cv2.VideoCapture("Traffic.mp4")

# Check video opened or not
if not cap.isOpened():
    print("Error opening video")
    exit()

while True:

    # Read frame from video
    ret, frame = cap.read()

    # If video ends
    if not ret:
        print("Video completed")
        break

    # Run YOLO detection
    results = model(frame, imgsz=320)

    # Draw detections
    annotated_frame = results[0].plot()

    # Resize window frame (optional)
    annotated_frame = cv2.resize(annotated_frame, (1000, 700))

    # Show output video
    cv2.imshow("YOLOv8 Object Detection", annotated_frame)

    # Press Q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
